```python
from bs4 import BeautifulSoup
import requests
```

```python
#Q1 final block
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
soup = BeautifulSoup(page.content, "html")
soup.prettify()
header_tags=soup.find_all(['h2','h1','h3','h4','h5','h6'])
headers=[]
for i in header_tags:
    head=i.get_text()
    head=head.replace("\n","")
    headers.append(head)
headers
```

```python
#Q2 Final_block
page=requests.get('https://www.imdb.com/list/ls055592025/')
soup = BeautifulSoup(page.content, "html")
soup.prettify()
header_tags=soup.find_all('h3',class_="lister-item-header")
rating_tags=soup.find_all('div',class_='lister-item-content')
headers=[]
years=[]
ratings=[]
for i in header_tags:
    title=i.find('a')
    time=i.find('span',class_="lister-item-year text-muted unbold")
    name=title.get_text()
    year=time.get_text()
    year=year.replace('(',"")
    year=year.replace(')',"")
    #head=head.replace("\n","")
    headers.append(name)
    years.append(year)
    
for i in rating_tags:
    rating=i.find("span", class_="ipl-rating-star__rating").get_text()
    #head=head.replace("\n","")
    ratings.append(rating)
ratings
import pandas as pd
data_imdb={'Title':headers,'Year':years,'Rating':ratings}
data=pd.DataFrame(data_imdb)
data.to_csv('Top_100_IMDB.csv')
```

```python
#Q3 final_block
page=requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
soup = BeautifulSoup(page.content, "html")
soup.prettify()
header_tags=soup.find_all('td',class_="titleColumn")[0:100]
Titles=[i.find('a').get_text() for i in header_tags]
rating_tags=soup.find_all('td', class_='ratingColumn imdbRating')[0:100]
Ratings=[i.find('strong').get_text() for i in rating_tags]
years_tags=soup.find_all('span',class_="secondaryInfo")[0:100]
Years=[i.get_text() for i in years_tags]

import pandas as pd
data_imdb={'Title':Titles,'Year':Years,'Rating':Ratings}
data=pd.DataFrame(data_imdb)
data.to_csv('Top_100_IMDB_Indian.csv')
```

```python
#Q4 final block
page=requests.get('https://bookpage.com/')
soup = BeautifulSoup(page.content, "html")
soup.prettify()
link='https://bookpage.com/'+soup.find('a',class_="dropdown-toggle")['href']
page1=requests.get(link)
soup1=BeautifulSoup(page1.content,"html")
soup1.prettify()
info_tags=soup1.find_all('div',class_="flex-article-content")[0:5]
book_names=[]
author_names=[]
genre_names=[]
reviews=[]
for i in info_tags:
    h4=i.find('h4',class_="italic").find('a').get_text()
    link2='https://bookpage.com/'+i.find('h4',class_="italic").find('a')['href']
    page2=requests.get(link2)
    soup2=BeautifulSoup(page2.content,"html")
    rev_tags=soup2.find('div',class_="article-body").find_all('p')
    r_=""
    for l in rev_tags:
        n=l.get_text().replace('\n',"")
        r_=r_+n
    
    reviews.append(r_)    
        
    book_names.append(h4)
    p=i.find('p',class_="sans bold").get_text().replace("\n","")
    author_names.append(p)
    g=i.find('p',class_="genre-links hidden-phone").find_all('a')
    g_=""
    m=0
    for j in g:
        k=j.get_text()
        if m==0:
            g_=g_+k
        if m>0:
            g_=g_+", "+k
        m=m+1
    genre_names.append(g_)
data_books={'Book_Name':book_names,'Author_Name':author_names,'Genre':genre_names,'Review':reviews}
data=pd.DataFrame(data_books)
data.to_csv('Books.csv')
```

```python
#Q7
page=requests.get('https://www.amazon.in/s?k=mobile+under+20000&ref=nb_sb_noss_2')
soup = BeautifulSoup(page.content, "html")
soup.prettify()
mobile_tags=soup.find_all('span',class_="a-size-medium a-color-base a-text-normal")
Mobile=[i.get_text() for i in mobile_tags]
price_tags=soup.find_all('span',class_="a-price-whole")
Price=[i.get_text() for i in price_tags]
image_url_tags=soup.find_all('img',class_='s-image')
Image_URL=[i['src'] for i in image_url_tags]
rating_tags=soup.find_all('span', class_="a-icon-alt")
Rating=[i.get_text() for i in rating_tags]

data_amazon={'Mobile':Mobile,'Price':Price,'Image_URL':Image_URL,'Rating':Rating[4:]}
data=pd.DataFrame(data_amazon)
data.to_csv('Mobiles_Under_20000.csv')

```
